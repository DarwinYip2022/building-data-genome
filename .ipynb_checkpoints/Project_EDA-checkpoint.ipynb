```python
def convert_categorical_to_int(input_object):
    # Remove the rupee symbol and commas, then convert to integer
    string = str(input_object)
    return float(string.replace('₹', '').replace(',', '').replace('|','0').replace('%',''))
```

```python
numerical_features = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# List of numerical features to convert
numerical_features = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# Convert all numerical features using the provided function
for feature in numerical_features:
    df[feature] = df[feature].apply(convert_categorical_to_int)

# Verify the changes by checking the data types
df[numerical_features].dtypes

```

### Approach

Specifically, we tackles the 'category' column where each entry is a string containing multiple subcategories separated by a pipe symbol '|'. The process starts by identifying and extracting the first two subcategories from each entry to capture the primary and secondary category information.

Once these unique subcategories are identified, we will generate new features (columns) in the dataset for each subcategory. These features are binary, indicating the presence (1) or absence (0) of each subcategory within the first two positions of the original 'category' entry for each record. This binary encoding allows for the categorical data to be easily used in various later data analysis techniques, such as clustering or machine learning models, where numerical input is required.

The outcome is a DataFrame with expanded features, enabling a more granular analysis of the data based on categorized attributes. This method is particularly useful in customer segmentation, product categorization, and other applications where categorical distinctions are important.

```python
unique_subcategories = set()

# Update the set with only the first two subcategories from each 'category' entry
df_cleaned['category'].str.split('|').apply(lambda x: unique_subcategories.update(x[:2]))

# Create a binary column for each of the unique subcategories (only the first two)
# We'll use a prefix 'subcat_' for each new subcategory column
for subcategory in unique_subcategories:
    col_name = f'{subcategory}'  # Unique column name with prefix
    df_cleaned[col_name] = df_cleaned['category'].apply(lambda x: 1 if subcategory in x.split('|')[:2] else 0)
```

## Clustering - Hierarchical

### 1. Goal: Given the products detailed description, can we find the hierarchy of its categorization? In other words, what product category should be grouped together given their similarities. 

- Improved Product Discovery: By accurately categorizing products into a hierarchical structure, businesses can enhance the search and discovery process on their platforms. This makes it easier for customers to find what they're looking for, improving the overall shopping experience.

- Targeted Marketing: Understanding product hierarchies enables more precise segmentation for marketing campaigns. Products can be grouped into relevant categories and subcategories, allowing for tailored marketing messages that are more likely to resonate with specific customer segments.

- Inventory Management: A clear view of product categorization helps in better inventory management. Businesses can identify which categories are most popular or underrepresented, informing stock decisions and helping to optimize inventory levels.

Here’s the breakdown of the process:

1. **Feature Extraction**: You start by converting the text descriptions of products into a numerical format that can be processed by clustering algorithms. This is done using the TF-IDF Vectorizer, which transforms the `about_product` text into a matrix of TF-IDF features. TF-IDF highlights words that are more relevant to each product, thus preparing the data for further analysis.

2. **Hierarchical Clustering**: With the features extracted, we then apply hierarchical clustering to the TF-IDF matrix. This method groups products based on the similarity of their descriptions, forming a hierarchy from the most distinct groups to a single cluster containing all products. The Ward method is used here, which minimizes the variance within each cluster.

3. **Cluster Assignment**: After forming the hierarchical structure, you decide on a number of clusters to focus on (40 in this case) and assign each product to one of these clusters. This step is crucial for simplifying the complex hierarchy into a more manageable number of groups.

4. **Identifying Common Categories**: For each of the 40 clusters, you analyze the categories of the products within them to find the most common category. This involves looking at the last part of each product's category path (indicating the most specific category) and identifying which category is most frequent in each cluster.

5. **Visualization**: Finally, you visualize the hierarchical clustering results with a dendrogram, but with a twist. Instead of labeling each leaf with product names or IDs, you label them with the most common category found in each cluster. This provides a clear visual representation of how products are grouped based on their descriptions and the dominant category in each group.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.cluster.hierarchy import ward, dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
# Use TF-IDF Vectorizer to convert text to feature vectors
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                   min_df=0.2, stop_words='english',
                                   use_idf=True, ngram_range=(1,3))

tfidf_matrix = tfidf_vectorizer.fit_transform(df_cleaned['about_product'])

# Perform hierarchical clustering
linkage_matrix = ward(tfidf_matrix.toarray())  # Use ward clustering algorithm on TF-IDF matrix

from collections import Counter

# Assign products to clusters
max_clusters = 40  # Assuming we want to identify the most common category for these many clusters
clusters = fcluster(linkage_matrix, max_clusters, criterion='maxclust')

# Add cluster assignments to the DataFrame
df_cleaned['cluster'] = clusters

# Find the most common category within each cluster
most_common_categories = []
for i in range(1, max_clusters + 1):
    cluster_categories = df_cleaned[df_cleaned['cluster'] == i]['category'].values
    # Splitting the categories and counting the most common base category
    base_categories = [cat.split('|')[-1] for cat in cluster_categories]  # Consider the base category for simplicity
    most_common_category = Counter(base_categories).most_common(1)[0][0]
    most_common_categories.append(most_common_category)

# Map cluster numbers to their most common categories
cluster_to_category = {i+1: category for i, category in enumerate(most_common_categories)}

# Update labels for the dendrogram to show the most common category of each cluster
updated_labels = [cluster_to_category[cluster] for cluster in clusters]

# Now let's plot the dendrogram again with the updated labels
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, truncate_mode='lastp', p=max_clusters, orientation="top", 
           labels=updated_labels, leaf_font_size=10, show_contracted=True)
plt.tight_layout()
plt.title('Hierarchical Clustering Dendrogram (Truncated) with Common Categories')
plt.xlabel('Cluster or (Cluster size)')
plt.ylabel('Distance')
plt.show()

# Return the mapping of clusters to their most common categories for reference
cluster_to_category
```

### Key Insights


1. **Cluster Labels Results** : The cluster results, which map each cluster number to its most common category (e.g., {1: 'SmartTelevisions', 2: 'USBCables', ...}), give you insight into the dominant product types within each cluster. This allows you to understand the variety and distribution of product types in your dataset, highlighting prevalent categories like USBCables, which appear in multiple clusters, indicating a variety of contexts or subcategories within USB cables themselves.
2. **Cluster Hierachary**: We can see how each cluster is grouped together. For instance, examining the leftmost tree nodes, clusters with dominant categories such as Mice and Wireless USB Adapters merge first, then they join with USB Cables, and eventually with Smart Televisions. This indicates that, based on the product descriptions, the clustering algorithm suggests Mice and Wireless USB Adapters products are similar; USB Cables are akin to both Mice and Wireless USB Adapters products, and so on.

## K-means Clustering

For review content and titles, it would be more advantageous to conduct review segmentation. We could apply K-means clustering to categorize user reviews into distinct groups. This approach would aid in understanding user preferences and their correlation with rating scores and sentiment analysis. Additionally, such segmentation can uncover patterns in customer feedback that are not immediately apparent. It can also highlight specific features or issues that are common within groups of reviews, providing actionable insights for product refinement and customer service improvement. By analyzing these segments, businesses can also tailor their communication strategies to address the concerns and preferences of different customer segments more effectively.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Assuming df_cleaned is your cleaned DataFrame containing the 'review_content', 'rating', and 'rating_count' columns

# Preprocess text data (simplified version)
preprocessed_text = df_cleaned['review_content'].str.lower()

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_text)

# Standardize 'rating_count' and 'review_length'
scaler = StandardScaler()
rating_count_scaled = scaler.fit_transform(np.log(df_cleaned[['rating_count']]))
review_length_scaled = scaler.fit_transform(np.log(df_cleaned[['review_length']]))

# Combine TF-IDF features with 'rating_count' and 'review_length'
features = np.hstack((tfidf_matrix.toarray(), rating_count_scaled, review_length_scaled))

# Apply K-means clustering to categorize products
kmeans = KMeans(n_clusters=10, random_state=42)
product_categories = kmeans.fit_predict(features)

# Add the cluster labels to your dataframe
df_cleaned['review_category'] = product_categories

# Scatter plot between "rating" and "review_length" using the cluster assignments for color
plt.figure(figsize=(10, 8))
scatter = plt.scatter(np.log(df_cleaned['rating_count']), np.log(df_cleaned['review_length']), 
                      c=df_cleaned['review_category'], cmap='viridis', alpha=0.6)

# Create a legend
plt.legend(handles=scatter.legend_elements()[0], title="Review Categories")

# Adding title and labels
plt.title('Scatter Plot of Rating_votes vs. Review Length')
plt.xlabel('log_Rating_count')
plt.ylabel('log_Review Length')

# Show the plot
plt.show()
```

- The plot displays a wide dispersion of review lengths across different rating counts.

- Different colors represent the 10 different clusters/categories identified by the K-means algorithm. The clusters appear to be distributed across all ranges of rating counts and review lengths, suggesting that the clusters may represent different characteristics beyond these two features (**which is captured by the review content in the tf-idf**) .

- There is a weak positive correlation that can be deduced between review length and rating count from this plot alone. It does appear that longer reviews correlate with a higher up-vote.
